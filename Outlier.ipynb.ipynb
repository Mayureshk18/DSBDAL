```python
sorted_rsscore=sorted(df['reading_score'])
```

```python
sorted_rsscore
```

```python
q1=np.percentile(sorted_rsscore,25)
q3=np.percentile(sorted_rsscore,75)
```

```python
print(q1,q3)
```

```python
IQR=q3-q1
```

```python
lwr_bound=q1-(1.5*IQR)
upr_bound=q3+(1.5*IQR)
print(lwr_bound,upr_bound)
```

```python
r_outliers=[]
for i in sorted_rsscore:
    if(i<lwr_bound or i>upr_bound):
        r_outliers.append(i)
print(r_outliers)
```
