```python
length = float(input("Enter the length of the field in feet: "))
width = float(input("Enter the width of the field in feet: "))

area_sq_feet = length * width
area_acres = area_sq_feet / 43560

print("The area of the field is", area_acres, "acres")
```

```python
meal_cost = float(input("Enter cost of meal: "))

tax_rate = 0.05         
tip_rate = 0.18         

tax = meal_cost * tax_rate
tip = meal_cost * tip_rate
total = meal_cost + tax + tip

print("Tax amount: ₹%.2f" % tax)
print("Tip amount: ₹%.2f" % tip)
print("Total amount: ₹%.2f" % total)
```

```python
widgets = int(input("Enter the number of widgets: "))
gizmos = int(input("Enter the number of gizmos: "))

weight_widgets = widgets * 75
weight_gizmos = gizmos * 112

total_weight = weight_widgets + weight_gizmos

print("The total weight of the order is", total_weight, "grams")
```
